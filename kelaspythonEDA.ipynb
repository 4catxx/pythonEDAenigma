```python
df.drop('email', axis=1, inplace=True)
```

```python
stok = df['brand'].groupby(df['model']).count()
stoktop15 = stok.nsmallest(5)
stoktop15.head(5)
plt.bar(stoktop15.index, stoktop15)
plt.ylabel("nama produk")
plt.xlabel("jumlah")
plt.title("Jumlah Produk terjual")
plt.show()
```

```python
stokbrand = df.groupby(['brand'])['id'].count()
stoktop10 = stokbrand.nlargest(5)
stoktop15.head
plt.barh(stoktop10.index, stoktop10)
plt.ylabel("nama produk")
plt.xlabel("jumlah")
plt.title("Jumlah Produk terjual")
plt.savefig('coba.jpg')
plt.show()
```
